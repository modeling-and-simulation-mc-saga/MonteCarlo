```python
import matplotlib.pyplot as plt
import math
import sys
sys.path.append("../lib")
from pylib import pltInit,Point,DataFile
```

```python
def draw():
    plt.figure(figsize=(18,12),facecolor='white')
    plt.title('Random Spin Monte Carlo $N=256$')
    plt.xlabel('$t$')
    plt.ylabel('energy')
    for i in range(5):
        filename = f'RandomSpin-{i}.txt'
        xl = DataFile(filename).readData()
        tl = [k for k in range(len(xl))]
        plt.plot(tl,xl,linewidth=5)
    plt.savefig('RandomSpin.pdf')
```

```python
pltInit()
draw()
```
