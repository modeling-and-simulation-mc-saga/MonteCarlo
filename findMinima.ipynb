```python
import matplotlib.pyplot as plt
import math
import sys
sys.path.append("../lib")
from pylib import pltInit,Point,DataFile
```

```python
def draw():
    plt.figure(figsize=(18,12), facecolor='white')
    plt.title('Searching minima')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.xticks([])
    plt.yticks([])
    xmax = 11
    xmin = -11
    plt.xlim(xmin,xmax)
    plt.ylim(-1000,7000)
    n=1024
    xl = [(xmax-xmin)*x/n+xmin for x in range(n)]
    yl = [math.pow(x,4)-50*math.pow(x,2) for x in xl]
    plt.plot(xl,yl)
    for k in [1,2]:
        filename = f'findMinima{k}.txt'
        xl,yl=DataFile(filename).readData()
        plt.scatter(xl,yl,s=100,label=f'start from point {k}')
    plt.legend()
    plt.savefig('findMinima.pdf')
```

```python
pltInit()
draw()
```
